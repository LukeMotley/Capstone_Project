```python
# Dependencies
import pandas as pd
import yfinance as yf
from yahoofinancials import YahooFinancials
```

```python

# Start and End date Variables
start_time = '2020-12-1'
end_time = '2021-4-30'

# Remember to input Ticker symbol in all caps
MARA_df = yf.download('MARA', 
                      start_time, 
                      end_time, 
                      progress=False)
MARA_df
```

```python
SONY_df = yf.download('SONY', 
                      start_time, 
                      end_time, 
                      progress=False)
SONY_df
```

```python
EA_df = yf.download('EA',
                      start_time, 
                      end_time, 
                      progress=False)
EA_df
```

```python
AMC_df = yf.download('AMC',
                      start_time, 
                      end_time, 
                      progress=False)
AMC_df
```

```python
GME_df = yf.download('GME',
                      start_time, 
                      end_time, 
                      progress=False)
GME_df
```

```python
VOO_df = yf.download('VOO',
                      start_time, 
                      end_time, 
                      progress=False)
VOO_df
```

```python
MARA_df.to_csv("MARA_df.csv")
SONY_df.to_csv("SONY_df.csv")
EA_df.to_csv("EA_df.csv")
AMC_df.to_csv("AMC_df.csv")
GME_df.to_csv("GME_df.csv")
VOO_df.to_csv("VOO_df.csv")
```
